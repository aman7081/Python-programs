```python
'''Write a python code that reads a file named "temp.txt", processes each word in the
file, counts the occurrences of each character (ignoring case), and stores the counts
in a dictionary named ch.'''

ch=dict()

fname="temp.txt"
fh=open(fname,encoding='utf-8-sig')
for line in fh:
  line.rstrip()
  word=line.split()
  for j in range(len(word)):
    tempword=word[j].lower()
    for k in tempword:
      if(k in ch):
        ch[k]=ch[k]+1
      else:
        ch[k]=1
ch
```

```python
#Write a python program to display Fibonacci series using recursion.
def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))
# take input from the user
nterms = int(input("How many terms? "))
# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

```python
'''Given three sequences of numbers, write a lambda expression using map and
zip functions,that will calculate element-wise maximum among these 3
sequences. Write a code to find floor, ceil and round of a given number.'''
sequences_max = lambda seq1, seq2, seq3: list(map(lambda x: max(x), zip(seq1, seq2, seq3)))
seq1 = [1, 5, 3, 9]
seq2 = [2, 6, 8, 4]
seq3 = [7, 2, 9, 5]
result = sequences_max(seq1, seq2, seq3)
print(result)
```

```python
'''Write a Python program to create a file where all letters of English alphabet
are listed by specified number of letters on each line.'''
letters_per_line = 5
with open("alphabet.txt", "w") as file:
    current_letter = ord('a')  # Get the ASCII value of 'a'
    while current_letter <= ord('z'):
        line_letters = ""
        for _ in range(letters_per_line):
            if current_letter <= ord('z'):
                line_letters += chr(current_letter) + " "
                current_letter += 1
        file.write(line_letters.strip() + "\n")
a=open("/content/alphabet.txt",'r')
print(a.read())
```

```python
'''Write a program in python to input student details like student roll no, student
name and marks in 5 subjects etc. and calculate the student grade based on
student percentage and printing all details.'''
class student:
  def __init__(self,roll_no,name,marks):
    self.roll_no=roll_no
    self.name=name
    self.marks=marks
  def grade(self):
    total=sum(self.marks)
    percentage=(total/500)*100
    print("Percentage: ", percentage)
    if percentage>=90:
      return 'A'
    elif percentage>=80:
      return 'B'
    elif percentage>=65:
      return 'C'
    elif percentage>=50:
      return 'D'
    else:
      return "Fail"
  def display_info(self):
     print("\nStudent Details:")
     print("Roll No:", self.roll_no)
     print("Name:", self.name)
     print("Marks:", self.marks)
     print("Total Marks:", sum(self.marks))
     print("Grade: ",self.grade())

#main program
name=input("Enter the name of the student: ")
roll_no=int(input("Enter the roll no: "))
marks=[]
for i in range(5):
  sub_marks=int(input(f"Enter the marks of subject {i+1}: "))
  marks.append(sub_marks)
s=student(roll_no,name,marks)
s.display_info()
```

```python
'''14.Write a program to demonstrate single inheritance to calculate cube of a
number.'''
class cube:
  def __init__(self,num):
    self.num=num
  def ans(self):
    ans=self.num**3
    print("Cube of ",self.num," is ",ans)
class child(cube):
  pass
c=child(5)
c.ans()
```
